```python
# Importing all the necessary libraries

import json
import requests
import urllib
import time
user='admin'
password='Welcome?'
url='http://cacib.opensee.ninja:8080'
module='irrbbMetrics_Nullable'
```

```python
# Preparing the API cals for ingestion

r= requests.get(url+'/ping')
print(r.json())
r= requests.post(url+'/login', json={"identifier":user,"password":password})
token=r.json()['token']
print(token)
r = requests.get(url+'/modules',  headers={'Authorization': f'{token}'})
```

```python
# Getting filepath and json data. Please update the variable's value if the dim group file is not in the same folder with this script.

path = "C:/Users/NurlanAkbarli/Documents/osg323/"  #path
json_data=json.load(open(path+'situation_name_dim_group_small.json','rb'))
```

```python
print(json_data[0])
```

```python
# Declaring dictionaries to list dimgroups names and bodies as a string.
dim_group_name_list=[]
dim_group_list=[]
for data in json_data:
    dim_group_name_list.append(data["name"])
    dm_re = str(data).replace('\'', '"')
    dim_group_list.append(dm_re)
```

```python
# Displaying an element of the group dimension
print(dim_group_name_list[0])
print(dim_group_list[0])
```

```python
# Prepare API call for posting dim groups
APIcall = '/module/irrbbMetrics_Nullable/dimension/group/'

for index in range (0,len(dim_group_list)):
    APIcallDimGroup = APIcall + dim_group_name_list[index]
    r = requests.post(url+APIcallDimGroup, headers={'Authorization': f'{token}','Accept' : 'application/json', 'Content-Type':'application/json'},data=dim_group_list[index])
    print(r.text)
    time.sleep(30)
```

```python
# import the second json file
json_dataput=json.load(open(path+'situation_name_dim_group_small_PUT.json','rb'))
print(json_dataput)
```

```python
# Prepare API call to put dim groups
APIcall = '/module/irrbbMetrics_Nullable/dimension/group/'

for index in range (0,len(dim_group_list)):
    APIcallDimGroup = APIcall + dim_group_name_list[index]
    r = requests.put(url+APIcallDimGroup, headers={'Authorization': f'{token}','Accept' : 'application/json', 'Content-Type':'application/json'},data=dim_group_list[index])
    print(r.text)
    time.sleep(30)
```
